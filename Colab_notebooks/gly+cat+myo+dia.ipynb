# **Images Preprocessing**

```python
train_path = '/content/ocular_disease_dataset_2/train'
val_path = '/content/ocular_disease_dataset_2/val'
```

```python
from glob import glob
folders = glob('/content/ocular_disease_dataset_2/train/*')
folders
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale = 1./255,
                                   horizontal_flip=True,
                                   rotation_range = 20)
val_datagen = ImageDataGenerator(rescale = 1./255,
                                 horizontal_flip=True,
                                 rotation_range = 20)
```

```python
batch_size = 32
img_height = 224
def generate_generator_multiple(generator,dir1,batch_size, img_height,img_width):
    genX1 = generator.flow_from_directory(dir1,
                                          target_size = (img_height,img_width),
                                          class_mode = 'categorical',
                                          batch_size = batch_size,
                                          shuffle=True, 
                                          seed=7)

    while True:
            X1i = genX1.next()
            yield [X1i[0],X1i[0]], X1i[1] 
            
            
inputgenerator=generate_generator_multiple(train_datagen,
                                           dir1=train_path,
                                           batch_size=batch_size,
                                           img_height=img_height,
                                           img_width=img_height)       
     
testgenerator=generate_generator_multiple(val_datagen,
                                          dir1=val_path,
                                          batch_size=batch_size,
                                          img_height=img_height,
                                          img_width=img_height)              
```

```python
training_set = train_datagen.flow_from_directory(train_path,
                                                 target_size = (224, 224),
                                                 batch_size = 32,
                                                 shuffle = True,
                                                 class_mode = 'categorical')
```

```python
val_set = val_datagen.flow_from_directory(val_path,
                                          target_size = (224, 224),
                                          batch_size = 32,
                                          shuffle = True,
                                          class_mode = 'categorical')
```

```python
training_set.class_indices
```

# **Model Training**

```python
from tensorflow.keras.models import load_model
model_combined = load_model('/content/drive/MyDrive/ai_healthcare/gly+cat+myo+dbts_model.h5')
```

```python
model_combined.trainable = False
```

```python
model_combined.summary()
```

```python
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.layers import Flatten
from tensorflow.keras.models import Model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPool2D , Flatten, MaxPooling2D
combi = model_combined.output
combi = Dense(64, activation='relu',name = 'combi1')(combi)
predictions = Dense(8, activation='softmax',name = 'combi2')(combi)
new_model = Model(inputs=model_combined.input, outputs=predictions)
```

```python
new_model.summary()
```

```python
from tensorflow.keras.optimizers import Adam
new_model.compile(optimizer=Adam(learning_rate = 0.001),loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
epo = 5
r = new_model.fit(inputgenerator,
                validation_data=testgenerator,
                epochs=epo,
                steps_per_epoch=len(training_set),
                validation_steps = len(val_set))
```

```python
epo = 5
r1 = new_model.fit(inputgenerator,
                validation_data=testgenerator,
                epochs=epo,
                steps_per_epoch=len(training_set),
                validation_steps = len(val_set))
```

```python
new_model.save('/content/drive/MyDrive/ai_healthcare/8_class_model_ensem/8_class_model_ensemble_2.h5')
```

```python
import matplotlib.pyplot as plt
epochs = [i for i in range(10)]
fig , ax = plt.subplots(1,2)
train_acc = r.history['accuracy'] + r1.history['accuracy']
train_loss = r.history['loss']+r1.history['loss']
val_acc = r.history['val_accuracy']+r1.history['val_accuracy']
val_loss = r.history['val_loss']+r1.history['val_loss']
fig.set_size_inches(16,9)

ax[0].plot(epochs , train_acc , 'go-' , label = 'Training Accuracy')
ax[0].plot(epochs , val_acc , 'ro-' , label = 'Testing Accuracy')
ax[0].set_title('Training & Validation Accuracy')
ax[0].legend()
ax[0].set_xlabel("Epochs")
ax[0].set_ylabel("Accuracy")

ax[1].plot(epochs , train_loss , 'g-o' , label = 'Training Loss')
ax[1].plot(epochs , val_loss , 'r-o' , label = 'Testing Loss')
ax[1].set_title('Testing Accuracy & Loss')
ax[1].legend()
ax[1].set_xlabel("Epochs")
ax[1].set_ylabel("Loss")

plt.show()
```

# **Results**

```python
batch_size = 32
img_height = 224
def generate_generator_multiple(generator,dir1,batch_size, img_height,img_width):
    genX1 = generator.flow_from_directory(dir1,
                                          target_size = (img_height,img_width),
                                          class_mode = 'categorical',
                                          batch_size = batch_size,
                                          shuffle=False, 
                                          seed=7)

    while True:
            X1i = genX1.next()
            yield [X1i[0],X1i[0]], X1i[1] 
            
            
inputgenerator_eval=generate_generator_multiple(train_datagen,
                                           dir1=train_path,
                                           batch_size=batch_size,
                                           img_height=img_height,
                                           img_width=img_height)       
     
testgenerator_eval=generate_generator_multiple(val_datagen,
                                          dir1=val_path,
                                          batch_size=batch_size,
                                          img_height=img_height,
                                          img_width=img_height)              
```

```python
import numpy as np
Y_pred = new_model.predict_generator(testgenerator_eval, len(val_set.classes)//val_set.batch_size +1)
```

```python
y_pred = np.argmax(Y_pred, axis=1)
```

```python
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix
print('Confusion Matrix')
cf = confusion_matrix(val_set.classes, y_pred)
print(cf)
```

```python
print('Classification Report')
target_names = ['C','G','M','Mild_DR','Moderate_DR','N','Proliferate_DR','Severe_DR']
print(classification_report(val_set.classes, y_pred, target_names=target_names))
```
